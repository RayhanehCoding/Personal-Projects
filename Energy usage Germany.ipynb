```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/projects/time_series_15min_singleindex.csv'
data = pd.read_csv(file_path)
print(data.head())
print(data.isnull().sum())
print(data.info())
```

```python
print(data.columns)
print(data['utc_timestamp'].head())
print(data.dtypes)

data['utc_timestamp'] = pd.to_datetime(data['utc_timestamp'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['utc_timestamp'], data['DE_load_actual_entsoe_transparency'], label='Load (Germany)')
plt.xlabel('Timestamp')
plt.ylabel('Load (MW)')
plt.title('Energy Load Over Time')
plt.legend()
plt.show()
```
